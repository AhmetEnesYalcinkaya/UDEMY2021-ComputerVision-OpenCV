## Resimlerde Thresholding İşlemi

---

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Images/rainbow.jpg',0)
plt.imshow(img,cmap='gray')
```

```python
img.max()
```

```python
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh1,cmap='gray')
```

```python
img = cv2.imread('Images/crossword.jpg',0)
plt.imshow(img,cmap='gray')
```

*Matplotlib içerisinde foto ölçeğini büyütmek için yazdığımız fonksiyon*

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 =cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

*Farklı renk ve ışıkta **threshold** fonksiyonu herzaman en iyi değeri vermez. OPnun için **adaptive_threshold** kullanmamız gerekir. **Adaptive threshold** belli bir bölgeyi seçip oranın thresholdunu belirleyip resimin tamamına uygular. Böylelikle daha iyi sonuçlar elde ederiz.*

```python
img_1 = cv2.imread('Images/sudoku.jpg',0)
plt.imshow(img_1,cmap='gray')
```

```python
show_pic(img_1)
```

```python
ret, th1 =cv2.threshold(img_1,127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img_1,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
th2 = cv2.adaptiveThreshold(img_1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```
