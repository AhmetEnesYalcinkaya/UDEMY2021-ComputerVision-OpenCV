## Color Mappings

---

*Bir önceki bölümlerde **RGB** renk uzayında çalışmıştık. Ancak 1970 lerde geliştirilen başka renk uzaylarıda mevcut. **HSL(hue,saturation,lightness)**, **HSV(hue,saturation,value)** hue-renk, saturation-renksel doygunluk lightess-canlılık.**HSL** ve **HSV** geliştirilen bir başka renk modelleridir.*

**cvtColor** fonksiyonunu kullanarak renk uzayını değiştirme işlemleri.

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Images/dog.jpg')
```

```python
plt.imshow(img) # default color spaces BGR
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(img)
```
