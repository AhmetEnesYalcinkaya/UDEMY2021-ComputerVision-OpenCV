```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('Images/horse.jpg')
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('Images/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

red_bricks = cv2.imread('Images/bricks.jpg')
show_bricks = cv2.cvtColor(red_bricks, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)
```

```python
plt.imshow(show_rainbow)
```

```python
plt.imshow(show_bricks)
```

```python
#OPENCV BGR
hist_values = cv2.calcHist([red_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
hist_values.shape
```

```python
plt.plot(hist_values)
```

```python
hist_values = cv2.calcHist([dark_horse],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
img = red_bricks
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,250])
    
plt.title('HISTOGRAM FOR BRICKS')
```

## Histogram Eqaulization

```python
rainbow = cv2.imread('Images/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
img = rainbow
```

```python
img.shape
```

```python
img.shape[:2]
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
```

```python
mask[50:150,50:200] = 255
```

```python
plt.imshow(mask,cmap='gray')
```

```python
plt.imshow(show_rainbow)
```

```python
masked_img = cv2.bitwise_and(img,img,mask=mask)
```

```python
show_masked_img = cv2.bitwise_and(show_rainbow,show_rainbow,mask=mask)
```

```python
plt.imshow(show_masked_img)
```

```python
# B G R
hist_mask_values_red = cv2.calcHist([rainbow],channels=[2],mask=mask,histSize = [256],ranges=[0,256])
```

```python
hist_values_red = cv2.calcHist([rainbow],channels=[2],mask=None,histSize = [256],ranges=[0,256])
```

```python
plt.plot(hist_mask_values_red)
plt.title('Red Histogram For Masked Rainbow')
```

```python
plt.imshow(show_rainbow)
```

```python
plt.plot(hist_values_red)
plt.title('Red Histogram For Normal Rainbow')
```
