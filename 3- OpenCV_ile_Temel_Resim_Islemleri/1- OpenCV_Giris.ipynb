```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

*Çevirme işlemini gerçekleştirdikten sonra fotoğrafımızın orjinal haline erişebileceğiz.*

```python
new_img = cv2.resize(fix_img,(1800,500))
```

```python
plt.imshow(new_img)
```

*Dikkat edersek burdaki x ve y koordinatlarının yerleri değişti.*

*Eğer bir oran vererek boyutunu ayarlamak istersek aaşağıdaki gibi kodumuzu düzenleyebiliriz.*

```python
w_ratio = 0.5
h_ratio = 0.5 #orjinal yüksekliğin %50'si
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

*Fotoğrafımızı ters çevirmek istersek **flip** fonksiyonunu kullanabiliriz.*

```python
new_img =cv2.flip(fix_img,0)
plt.imshow(new_img)
```

```python
new_img =cv2.flip(fix_img,1)
plt.imshow(new_img)
```

```python
new_img =cv2.flip(fix_img,-1)
plt.imshow(new_img)
```

*Fotoğrafımızı dosyaya kaydetmek için **cv2.imwrite()** fonksiyonunu kullanabiliriz.*

```python
cv2.imwrite('yeni_kopek.jpg',fix_img)
```
